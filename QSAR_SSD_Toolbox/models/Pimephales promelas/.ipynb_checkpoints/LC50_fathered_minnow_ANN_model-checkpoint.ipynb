## Creating a neural network model for LC50 values for fathered minnow

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import boston_housing
from keras import optimizers
from keras import regularizers
from keras.models import load_model
```

```python
# load all LC50 values for fathered minnows
import pandas as pd
import numpy as np

master_data = pd.read_excel('./data/LC50_pimephales_promelas_in_use.xlsx')
master_data_descs = pd.read_excel('./data/LC50_pimephales_promelas_in_use_selected_descriptors.xlsx', header=0)
```

### Load and split data

```python
from sklearn.model_selection import train_test_split

X = master_data_descs.values
y = master_data['Value'].values

X_trn, X_tst, y_trn, y_tst = train_test_split(X, y, test_size=0.2, random_state=3)
```

#### data preprocessing
Standardization

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# pca = PCA(n_components=60).fit(X_trn)
# X_trn = pca.transform(X_trn)
# X_tst = pca.transform(X_tst)

scaler = StandardScaler().fit(X_trn)
X_trn = scaler.transform(X_trn)
X_tst = scaler.transform(X_tst)
```

### Build ANN model

```python
# build a simple NN model
model = Sequential()

model.add(Dense(units=64, 
                kernel_initializer='TruncatedNormal', 
                kernel_regularizer=regularizers.l2(0.01),
                activation='relu', 
                input_dim=X_trn.shape[1]))

model.add(Dense(units=32, 
                kernel_regularizer=regularizers.l2(0.01),
                kernel_initializer='TruncatedNormal', 
                activation='relu'))
model.add(Dense(1)) # output layer

opt = optimizers.Adagrad(lr=0.01, decay=1e-6)

model.compile(optimizer= opt,
             loss='mse',
             metrics=['mae'])

history = model.fit(X_trn, y_trn, batch_size=32, epochs=300, verbose=1, validation_data=(X_tst, y_tst))
```

```python
# evaluation
from sklearn.metrics import r2_score

pred_trn = model.predict(X_trn)
pred_tst = model.predict(X_tst)
print r2_score(pred_tst, y_tst)
```

```python
# plot training processes
import matplotlib.pyplot as plt

plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])

plt.show()
```

```python
# plot
import matplotlib.pyplot as plt
upper_lim = max(max(y_trn), max(y_tst))

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

plt.plot(pred_trn, y_trn, 'o', label='training chemicals')
plt.plot(pred_tst, y_tst, '^', label='testing chemicals')

thisLine = plt.plot(np.append(0,upper_lim), np.append(0,upper_lim), label='1:1 ratio line')

plt.ylim([0, upper_lim])
plt.xlim([0, upper_lim])
plt.grid()
plt.legend(loc='upper left',numpoints=1)
plt.show()
```

```python
fig.savefig('./results/LC50_predicton_fathered_minnow_ANN.png')
```

```python
# save the ANN model
model.save('./results/LC50_fatered_minnow_ANN.h5')
```
